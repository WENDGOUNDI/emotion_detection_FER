```python
from fer import FER as f
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def emotion_dection(image):
    image = cv2.imread(image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    emotion_detector = f(mtcnn=True) # use MTCNN for face detection, otherwise, it will use cascade classifier as default
    get_emotions = emotion_detector.detect_emotions(image)
    emotions_dic = get_emotions[0]["emotions"]
    print(f"He/she is likely to be {max(emotions_dic, key=emotions_dic.get)}.")

    plt.imshow(image)
    plt.tick_params(left = False, right = False , labelleft = False ,
                    labelbottom = False, bottom = False)
    plt.show()
    
    
emotion_dection("./data/emotion_data/sad.jpg")
```
